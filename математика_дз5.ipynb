### Задание 2


Пусть брошены 3 уравновешенные монеты.

Рассмотрим 3 события:

A1 - монеты 1 и 2 упали одной и той же стороной;

A2 - монеты 2 и 3 упали одной и той же стороной;

A3 - монеты 1 и 3 упали одной и той же стороной.

Покажите, почему эти 3 события (A1, A2, A3) являются попарно независимыми, но не являются независимыми в совокупности.

```python
# При подбрасывании 3 монет общее количество исходов 2**3=8. Для любого из событий А1, А2, А3 количество успешных исходов 
# равно 4, т.е. р(А1)=р(А2)=р(А3)=4/8=0,5. Пересечение любых 2 событий (например, А1 и А2) будет не пустым множеством, если
# все 3 монеты упали одной стороной. Таких исходов 2. Тогда р(А1*А2)=р(А2*А3)=р(А1*А3)=р(А1*А2*А3)=2/8=0,25.
# По определению для независимых событий выполняется равенство р(А*В)=р(А)*р(В), отсюда события являются попарно 
# независимыми (0,25 = 0,5*0,5) и не являются независимыми в совокупности (0,25 != 0,5*0,5*0,5).
```

### Задание 3


Изучите 3 распределения из библиотеки scipy: [Нормальное](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.norm.html), [Экспоненциальное](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.expon.html), [Стьюдента](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.t.html). Используя функцию плотности распредления (`pdf`), постройте пять графиков плотностей для каждого распределения при разных параметрах. Запишите ваши наблюдения: как и на что влияют параметры у каждого распределения. 

Дополнительная информация [тут](https://pythonguides.com/scipy-stats/).

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
norm_x = np.linspace(-4,4,200)
pdf_norm = stats.norm.pdf(norm_x,loc=0,scale=1)
```

```python
plt.plot(norm_x,pdf_norm)
plt.title("Нормальное распределение - значения по умолчанию")
plt.show()
```

```python
pdf_norm = stats.norm.pdf(norm_x,loc=2,scale=1)
plt.plot(norm_x,pdf_norm)
plt.title("Нормальное распределение - среднее=2")
plt.show()
```

```python
pdf_norm = stats.norm.pdf(norm_x,loc=-1,scale=1)
plt.plot(norm_x,pdf_norm)
plt.title("Нормальное распределение - среднее=-1")
plt.show()
```

```python
pdf_norm = stats.norm.pdf(norm_x,loc=0,scale=2)
plt.plot(norm_x,pdf_norm)
plt.title("Нормальное распределение - стандартное отклонение=2")
plt.show()
```

```python
pdf_norm = stats.norm.pdf(norm_x,loc=0,scale=0.5)
plt.plot(norm_x,pdf_norm)
plt.title("Нормальное распределение - стандартное отклонение=0,5")
plt.show()
```

Изменение среднего значения приводит к сдвигу функции вдоль оси Х. Увеличение стандартного отклонения увеличивает хвосты функции и делает ее ниже и шире, уменьшение - выше и уже.

```python
from scipy.stats import expon
array_data = np.arange(-1,30,0.1)
```

```python
expon_pdf = expon.pdf(array_data,0,1)
plt.plot(array_data,expon_pdf,'bo')
plt.title("Экспоненциальное распределение - значения по умолчанию")
plt.show()
```

```python
expon_pdf = expon.pdf(array_data,0,5)
plt.plot(array_data,expon_pdf,'bo')
plt.title("Экспоненциальное распределение - стандартное отклонение=5")
plt.show()
```

```python
expon_pdf = expon.pdf(array_data,0,0.25)
plt.plot(array_data,expon_pdf,'bo')
plt.title("Экспоненциальное распределение - стандартное отклонение=0.25")
plt.show()
```

```python
expon_pdf = expon.pdf(array_data,5,1)
plt.plot(array_data,expon_pdf,'bo')
plt.title("Экспоненциальное распределение - среднее=5")
plt.show()
```

```python
expon_pdf = expon.pdf(array_data,0.2,1)
plt.plot(array_data,expon_pdf,'bo')
plt.title("Экспоненциальное распределение - среднее=0.2")
plt.show()
```

Увеличение среднего значения приводит к сдвигу вправо. Увеличение стандартного отклонения делает функцию более пологой, уменьшение - более крутой.

```python
from scipy.stats import t
df = 2.74
#mean, var, skew, kurt = t.stats(df, moments='mvsk')
```

```python
x = np.linspace(t.ppf(0.01, df),t.ppf(0.99, df), 100)
```

```python
t_pdf = t.pdf(x,df)
plt.plot(x,t_pdf)
plt.title("Распределение Стьюдента - по умолчанию")
plt.show()
```

```python
t_pdf = t.pdf(x,df,loc=2)
plt.plot(x,t_pdf)
plt.title("Распределение Стьюдента - среднее=2")
plt.show()
```

```python
t_pdf = t.pdf(x,df,loc=-1)
plt.plot(x,t_pdf)
plt.title("Распределение Стьюдента - среднее=-1")
plt.show()
```

```python
t_pdf = t.pdf(x,df,scale=2)
plt.plot(x,t_pdf)
plt.title("Распределение Стьюдента - стандартное отклонение=2")
plt.show()
```

```python
t_pdf = t.pdf(x,df,scale=0.2)
plt.plot(x,t_pdf)
plt.title("Распределение Стьюдента - стандартное отклонение=0.2")
plt.show()
```

```python
Функция ведет себя аналогично нормальному распределению.
Изменение среднего значения приводит к сдвигу функции вдоль оси Х. 
Увеличение стандартного отклонения увеличивает хвосты функции и делает ее ниже и шире, уменьшение - выше и уже.
```
