### Задание 1

Дан объект в $2D$ пространстве

````python
import numpy as np
import matplotlib.pyplot as plt

A = [
[-100, -100, 1],
[-100, 100, 1],
[100, 100, 1],
[0, 0, 1],
[100, -100, 1],
[-100, -100, 1]
]

A = np.array(A)

x = A[:,0]
y = A[:,1]
plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200]) 
plt.show()
````

При помощи линейных отображений:
- Уменьшить объект в два раза
- Повернуть на 130 градусов
- Отразить объект относительно прямой y=x (поможет матрица `[[0, 1, 0], [1, 0, 0], [0, 0, 1]]`)


````python
B = [[0.5,0,0],
     [0,0.5,0],
     [0,0,1]]
A1 = A@B
x = A1[:,0]
y = A1[:,1]
plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200]) 
plt.show()
````

````python
u = np.deg2rad(130)
B1 = [[np.cos(u),np.sin(u),0],
      [-np.sin(u),np.cos(u),0],
      [0,0,1]]
A2 = A1@B1
x = A2[:,0]
y = A2[:,1]
plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200]) 
plt.show()
````

````python
B2 = [[0, 1, 0], 
      [1, 0, 0], 
      [0, 0, 1]]
A3 = A2@B2
x = A3[:,0]
y = A3[:,1]
plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200]) 
plt.show()
````

### Задание 2

Найдите спектральное разложение матрицы:
```
m = np.array([[1, 2],
              [2, 3]])
print(m)
```



````python
m = np.array([[1, 2],
              [2, 3]])
print(m)
````

````python
res = np.linalg.eig(m)
res1 = np.transpose(res[1])
print('число 1:', res[0][0], ', вектор 1:', res1[0]/res1[0][0])
print('число 2:', res[0][1], ', вектор 2:', res1[1]/res1[1][1])
````

````python
val1 = res[0][0]
val2 = res[0][1]
vec1 = res1[0]/res1[0][0]
vec2 = res1[1]/res1[1][1]
St = np.array([vec1,vec2])
S = np.transpose(St)
D = np.array([[val1,0],[0,val2]])
print('спектральное разложение матрицы m = S @ D @ St')
print('S=',S)
print('D=',D)
print('St=',St)
````
